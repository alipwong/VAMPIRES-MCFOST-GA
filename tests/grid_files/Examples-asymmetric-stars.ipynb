```python
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

## Working out the coordinate system

Changing the x, y and z offset of 2
amp: 10
sigma: 0.1

```python
def extract_img(directory):
    img = fits.open("{}/RT.fits".format(directory))[0].data
    I = img[0, :, :, :, :]
    Q = img[1, :, :, :, :]
    U = img[2, :, :, :, :]
    V = img[3, :, :, :, :]
    P = np.divide(np.power(np.power(Q, 2) + np.power(U, 2) + np.power(V, 2), 0.5),I)

    imgs = [('I', I), ('Q', Q), ('U', U), ('P', P)]
    return imgs
```

```python
def plot_imgs(imgs):
    x1 = 475
    x2 = 525
    plt.figure(figsize=(12, 2.5))
    for j in range(4):
        plt.subplot(1, 4, j + 1)
        img = imgs[j][1]
        plt.imshow(img[0][0][x1:x2, x1:x2])
        plt.colorbar()
```

### Changing the x offset

```python
# Naming system: axis, offset, inc, az
imgs = extract_img("x_2_0_0")
plot_imgs(imgs)
imgs = extract_img("x_2_90_0")
plot_imgs(imgs)
imgs = extract_img("x_2_0_90")
plot_imgs(imgs)
imgs = extract_img("x_2_90_90")
plot_imgs(imgs)
```
